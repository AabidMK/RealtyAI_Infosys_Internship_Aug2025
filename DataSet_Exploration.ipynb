```
import pandas as pd
import re
import spacy
nlp = spacy.load("en_core_web_sm")
```

```
# Read your dataset
indian_cities = pd.read_csv(
    "Real Estate Data V21.csv"
)
# Quick look at the data
indian_cities.head()
```

```
# Mapping for text-based numbers
TEXT_NUMBERS = {
    "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
    "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10
}

# Regex patterns for common cases
ROOM_PATTERNS = [
    r'(\d+)\s*bhk',
    r'(\d+)\s*bed(?:room)?\b',
    r'(\d+)\s*br\b',
    r'(\d+)\s*rk\b'
]

def extract_rooms(text):
    if not isinstance(text, str):
        return None
    text_lower = text.lower()

    # 1. Regex direct match
    for pat in ROOM_PATTERNS:
        match = re.search(pat, text_lower)
        if match:
            return int(match.group(1))

    # 2. Spelled-out numbers
    doc = nlp(text_lower)
    for i, token in enumerate(doc):
        if token.text in TEXT_NUMBERS:
            if i + 1 < len(doc):
                next_word = doc[i + 1].text
                if next_word in ["bhk", "bedroom", "bed", "br", "rk"]:
                    return TEXT_NUMBERS[token.text]

    # 3. Fallback: numeric token + room word
    for tok, nxt in zip(doc, doc[1:]):
        if tok.like_num and nxt.text in ["bhk", "bedroom", "bed", "br", "rk"]:
            try:
                return int(tok.text)
            except ValueError:
                pass

    return None
```

```
print(indian_cities.columns.tolist())
indian_cities["num_rooms"] = indian_cities["Description"].apply(extract_rooms)
indian_cities[["Description", "num_rooms"]].head(20)
```

```
print("Total entries:", len(indian_cities))
print("Extracted room counts:", indian_cities['num_rooms'].notnull().sum())
print("Missing room counts:", indian_cities['num_rooms'].isnull().sum())
```

```
indian_cities.to_csv(
    "Real Estate Data V21_with_rooms.csv",
    index=False
)
print("✅ File saved with new column 'num_rooms'")
```
