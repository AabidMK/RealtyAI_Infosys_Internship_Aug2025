```python
import pandas as pd
df = pd.read_csv("State_time_series.csv")
df.head()

```

```python
# Step 1 â€” check data types
df.info()

# Step 2 â€” convert Date column to datetime if it's object type
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Step 3 â€” confirm conversion worked
print(df['Date'].dtypes)
print(df['Date'].head())
```

```python
mv = df[['Date', 'RegionName', 'ZHVI_AllHomes']].copy()

# Sort by RegionName and Date for clean time series
mv = mv.sort_values(['RegionName', 'Date']).reset_index(drop=True)

mv.head()
```

```python
mv['ZHVI_AllHomes'] = mv.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.interpolate(method='linear'))
```

```python
mv['ZHVI_AllHomes'] = mv.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.ffill().bfill())
```

```python
mv['ZHVI_AllHomes'].isna().sum()
```

```python
# Fill any remaining missing values using forward fill + backward fill
mv['ZHVI_AllHomes'] = mv.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.ffill().bfill())

# Check again to confirm there are no NaN values now
missing_after_fill = mv['ZHVI_AllHomes'].isna().sum()
print(f" Missing values after forward/backward fill: {missing_after_fill}")
```

```python
# Find which RegionNames still have missing ZHVI_AllHomes
missing_regions = mv[mv['ZHVI_AllHomes'].isna()]['RegionName'].unique()
print("Regions with remaining missing values:")
print(missing_regions)

# Check how many rows are missing for each region
missing_counts = mv[mv['ZHVI_AllHomes'].isna()]['RegionName'].value_counts()
print("\nMissing value count per region:")
print(missing_counts.head(10))
```

```python
# Forward and backward fill region-wise using transform
mv['ZHVI_AllHomes'] = mv.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.ffill().bfill())

# Check remaining missing values
remaining_mv = mv['ZHVI_AllHomes'].isna().sum()
print(f"Missing values after forward/backward fill: {remaining_mv}")
```

```python
# Drop rows where RegionName is UnitedStates
mv = mv[mv['RegionName'] != 'UnitedStates']

# Forward/backward fill again for Louisiana
mv['ZHVI_AllHomes'] = mv.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.ffill().bfill())

# Check missing values now
remaining_mv = mv['ZHVI_AllHomes'].isna().sum()
print(f"Missing values after cleaning: {remaining_mv}")
```

```python
# Make a clean copy without UnitedStates
mv_clean = mv[mv['RegionName'] != 'UnitedStates'].copy()

# Forward/backward fill for each region
mv_clean['ZHVI_AllHomes'] = mv_clean.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.ffill().bfill())

# Check remaining missing values
remaining_mv = mv_clean['ZHVI_AllHomes'].isna().sum()
print(f"Missing values after cleaning: {remaining_mv}")
```

```python
median_value = mv_clean[mv_clean['RegionName'] == 'Louisiana']['ZHVI_AllHomes'].median()
mv_clean.loc[mv_clean['RegionName'] == 'Louisiana', 'ZHVI_AllHomes'] = mv_clean.loc[mv_clean['RegionName'] == 'Louisiana', 'ZHVI_AllHomes'].fillna(median_value)
```

```python
mv_clean = mv_clean.dropna(subset=['ZHVI_AllHomes'])
```

```python
print(mv_clean['ZHVI_AllHomes'].isna().sum())
```

```python
# 1. Convert 'Date' to datetime format
mv_clean['Date'] = pd.to_datetime(mv_clean['Date'])

# 2. Sort by RegionName and Date (important for time series)
mv_clean = mv_clean.sort_values(by=['RegionName', 'Date'])

# 3. Optional: set 'Date' as index if you want time-series operations
mv_clean.set_index('Date', inplace=True)

# 4. Quick check
print(mv_clean.head())
print(mv_clean.info())
```

```python
import matplotlib.pyplot as plt

# Select a few regions to visualize
regions_to_plot = ['Alabama', 'Louisiana', 'California']

plt.figure(figsize=(12,6))

for region in regions_to_plot:
    mv_clean_region = mv_clean[mv_clean['RegionName'] == region]
    plt.plot(mv_clean_region.index, mv_clean_region['ZHVI_AllHomes'], label=region)

plt.title('ZHVI Trends Over Time')
plt.xlabel('Date')
plt.ylabel('ZHVI All Homes')
plt.legend()
plt.show()
```

```python
df_multivariate = mv_clean.reset_index()[['Date', 'RegionName', 'ZHVI_AllHomes']].copy()
df_multivariate.head()
```

```python
# Convert 'Date' to index temporarily for rolling operation (if not already)
df_multivariate.set_index('Date', inplace=True)

# Function to correct outliers using rolling median
def correct_outliers(series, window=5):
    rolling_median = series.rolling(window=window, center=True).median()
    diff = abs(series - rolling_median)
    threshold = 1.5 * diff.median()
    series_corrected = series.copy()
    series_corrected[diff > threshold] = rolling_median[diff > threshold]
    return series_corrected

# Apply outlier correction region-wise
df_multivariate['ZHVI_AllHomes'] = df_multivariate.groupby('RegionName')['ZHVI_AllHomes'].transform(correct_outliers)

# Reset index if needed
df_multivariate.reset_index(inplace=True)

# Quick check
df_multivariate.head()
```

```python
# Filter for Alabama before correction
alabama_before = df_multivariate[df_multivariate['RegionName'] == 'Alabama']['ZHVI_AllHomes'].describe()
print("Before correction:\n", alabama_before)

# Apply rolling median correction separately for checking
def correct_outliers(series, window=5):
    rolling_median = series.rolling(window=window, center=True).median()
    diff = abs(series - rolling_median)
    threshold = 1.5 * diff.median()
    series_corrected = series.copy()
    series_corrected[diff > threshold] = rolling_median[diff > threshold]
    return series_corrected

alabama_after = correct_outliers(df_multivariate[df_multivariate['RegionName'] == 'Alabama']['ZHVI_AllHomes'])
print("\nAfter correction:\n", alabama_after.describe())
```

```python
df_multivariate['ZHVI_RollingMean'] = df_multivariate.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.rolling(window=12).mean())
```

```python
# Show first few rows including the rolling mean
df_multivariate[['Date', 'RegionName', 'ZHVI_AllHomes', 'ZHVI_RollingMean']].head(20)
```

```python
import matplotlib.pyplot as plt

# Pick a region to visualize
region = 'Alabama'
region_data = df_multivariate[df_multivariate['RegionName'] == region]

plt.figure(figsize=(12,6))
plt.plot(region_data['Date'], region_data['ZHVI_AllHomes'], label='Original ZHVI', alpha=0.7)
plt.plot(region_data['Date'], region_data['ZHVI_RollingMean'], label='12-Month Rolling Mean', color='red', linewidth=2)
plt.title(f'ZHVI Trends with Rolling Mean - {region}')
plt.xlabel('Date')
plt.ylabel('ZHVI All Homes')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load dataset
mv = pd.read_csv("State_time_series.csv")

# Convert Date column to datetime
mv['Date'] = pd.to_datetime(mv['Date'])

# Drop missing values in ZHVI_AllHomes
mv = mv.dropna(subset=['ZHVI_AllHomes'])

# Convert Date to numeric (ordinal) for model training
mv['Date_ordinal'] = mv['Date'].map(pd.Timestamp.toordinal)
```

```python
from sklearn.linear_model import LinearRegression
import joblib, os

os.makedirs("models", exist_ok=True)  # Create folder if not exists

regions = mv['RegionName'].unique()

for region in regions:
    region_data = mv[mv['RegionName'] == region]

    # Skip small regions with too few data points
    if len(region_data) < 10:
        continue

    X = region_data[['Date_ordinal']]
    y = region_data['ZHVI_AllHomes']

    model = LinearRegression()
    model.fit(X, y)

    # Save each model
    joblib.dump(model, f"models/model_{region}.joblib")
    print(f" Model saved for region: {region}")

print(" All regional models trained and saved successfully!")
```

```python
import matplotlib.pyplot as plt

for region in mv['RegionName'].unique():
    region_data = mv[mv['RegionName'] == region]

    if len(region_data) < 10:
        continue

    plt.figure(figsize=(10,5))
    plt.plot(region_data['Date'], region_data['ZHVI_AllHomes'], label='Original ZHVI', alpha=0.7)
    plt.title(f'ZHVI Trends - {region}')
    plt.xlabel('Date')
    plt.ylabel('ZHVI All Homes')
    plt.legend()
    plt.show()
```

```python
import pandas as pd

# Load CSV
df = pd.read_csv("State_time_series.csv")

# Remove leading/trailing spaces in column names
df.columns = df.columns.str.strip()

# Check first few rows
print(df.head())
```

```python
df['ZHVI_AllHomes'] = pd.to_numeric(df['ZHVI_AllHomes'], errors='coerce')
df['ZHVI_AllHomes'].fillna(method='ffill', inplace=True)
df['ZHVI_AllHomes'].fillna(method='bfill', inplace=True)
```

```python
# Fill missing values forward, then backward
df['ZHVI_AllHomes'] = df['ZHVI_AllHomes'].ffill().bfill()
```

```python
# Check first 10 rows
print(df[['Date', 'RegionName', 'ZHVI_AllHomes']].head(10))

# Check if there are still any NaNs
print(df['ZHVI_AllHomes'].isna().sum())
```

```python
# 12-month rolling mean for each region
df['ZHVI_RollingMean'] = df.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.rolling(window=12).mean())
```

```python
df['Date_ordinal'] = pd.to_datetime(df['Date']).map(pd.Timestamp.toordinal)
```

```python
from sklearn.linear_model import LinearRegression
import joblib
import os

os.makedirs('models', exist_ok=True)

for region in df['RegionName'].unique():
    region_data = df[df['RegionName'] == region]
    X = region_data[['Date_ordinal']]
    y = region_data['ZHVI_AllHomes']

    model = LinearRegression()
    model.fit(X, y)

    joblib.dump(model, f'models/{region}_model.pkl')
    print(f'Model saved for region: {region}')
```

```python
import streamlit as st
import pandas as pd
import joblib
import numpy as np
import os

# Load your dataset
df = pd.read_csv("State_time_series.csv")

# Fill missing ZHVI values
df['ZHVI_AllHomes'] = df['ZHVI_AllHomes'].ffill().bfill()
df['ZHVI_RollingMean'] = df.groupby('RegionName')['ZHVI_AllHomes'].transform(lambda x: x.rolling(window=12).mean())
df['Date_ordinal'] = pd.to_datetime(df['Date']).map(pd.Timestamp.toordinal)

# App title
st.title("ZHVI Price Prediction & Forecasting")

# Tabs for Prediction and Forecast
tab1, tab2 = st.tabs(["Price Prediction", "Forecasting"])

# --- Price Prediction Tab ---
with tab1:
    st.header("Predict ZHVI for a region")
    region = st.selectbox("Select Region", df['RegionName'].unique())
    date_input = st.date_input("Select Date")

    # Load model
    model_path = f"models/{region}_model.pkl"
    model = joblib.load(model_path)

    if st.button("Predict Price"):
        date_ord = pd.Timestamp(date_input).toordinal()
        pred = model.predict([[date_ord]])
        st.success(f"Predicted ZHVI: ${pred[0]:,.2f}")

# --- Forecasting Tab ---
with tab2:
    st.header("ZHVI Rolling Mean Forecast")
    region_f = st.selectbox("Select Region for Forecast", df['RegionName'].unique(), key="forecast")

    region_data = df[df['RegionName'] == region_f]
    st.line_chart(region_data[['ZHVI_AllHomes', 'ZHVI_RollingMean']].set_index(region_data['Date']))
```
