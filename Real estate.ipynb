```python
import pandas as pd

# Load CSV
df = pd.read_csv(r"C:\Users\vansh\Downloads\archive (4)\Real Estate Data V21.csv")
```

```python
# we used re-regular expression (regex module) so that we can find the patterns such as 3bhk and 5 bhk so on..
```

```python
import re

# we used astype to convert the column values into string and then from string values we converted it into lowercase
# and then stored it in separate column name called property_title.
df['property_title'] = df['Property Title'].astype(str).str.lower()

def extract_rooms(text):
    # it extracts the numbers like 3 and 4 and for identifying bhk we used *bhk
    match = re.search(r'(\d+(\.\d+)?)\s*bhk', text)
    if match:
        return float(match.group(1))

    # if there is any description like bedrooms we use below function
    match = re.search(r'(\d+)\s*bed(room)?', text)
    if match:
        return int(match.group(1))

    # if description is present such as rk using below function
    match = re.search(r'(\d+)\s*rk', text)
    if match:
        return int(match.group(1))

    return None

# stores no of rooms in new column called num_rooms
df['num_rooms'] = df['property_title'].apply(extract_rooms)
print(df[['property_title', 'num_rooms']].head(20))
```
